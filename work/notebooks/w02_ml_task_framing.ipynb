## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A single hand rule (`stale x visible x impressions`) is transparent and even ties a tiny tree at the top of the list, but the pattern is *messy*: many tangled signals (impressions, position, CTR, content age, staleness) shift across clients and over time. On held-out clients the depth-2 tree matched the rule, yet the random forest still beat it **0.68 vs 0.24** — ML earns its place where one if-statement runs out of signal and can't adapt per client.
